### 5.1 로그 변환

로그 변환은 정규성을 높이고 분석에서 정확한 값을 얻기 위해 필요합니다. 이때 log는 큰 수를 같은 비율의 작은 수로 바꿔주는 역할을 합니다. 

로그 변환은 **np.log1p**를 이용하면 되는데요! 아래 코드처럼 로그 변환하고 싶은 열을 지정해 로그 변환을 진행하면 아주 쉽죠?

```python
train_x['log_Flight Distance'] = np.log1p(train_x['Flight Distance'])

test_x['log_Flight Distance'] = np.log1p(test['Flight Distance'])
```

그래프로 로그 변환 전, 후 분포를 확인해볼게요!

seaborn의 distplot을 사용하면 분포를 한눈에 확인할 수 있습니다.

```python
plt.figure(figsize = (8, 10))

plt.subplot(2,2,1)
sns.distplot(train_x['Flight Distance'])
plt.title('befor_log (train)')

plt.subplot(2,2,2)
sns.distplot(train_x['log_Flight Distance'])
plt.title('after_log (train)')

plt.subplot(2,2,3)
sns.distplot(test_x['Flight Distance'])
plt.title('befor_log (test)')

plt.subplot(2,2,4)
sns.distplot(test_x['log_Flight Distance'])
plt.title('after_log (test)')
```

확실히 로그 변환 후 분포가 달라진 것을 확인할 수 있습니다.
