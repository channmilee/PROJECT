# 5.예측값 대체

```python
train_xx = train_x.copy()
```

```python
train_xx.drop(columns = 'PC_L', inplace = True)
```

```python
from sklearn.preprocessing import LabelEncoder

#qualitative to quantitative
cols = ['PRODUCT_CODE', 'LINE']

for col in cols:
  le = LabelEncoder()
  le = le.fit(train_xx[col])
  train_xx[col] = le.transform(train_xx[col])

  for label in np.unique(test_x[col]):
    if label not in le.classes_:
      le.classes_ = np.append(le.classes_, label)
  test_x[col] = le.transform(test_x[col])

print('Done')
```

```python
train_a_le = train_xx[train_xx['PRODUCT_CODE'] == 0]
train_t_le = train_xx[train_xx['PRODUCT_CODE'] == 2]
train_o_le = train_xx[train_xx['PRODUCT_CODE'] == 1]
```

```python
train_a_le
```

```python
from sklearn.impute import SimpleImputer
imp_mean = SimpleImputer(strategy = 'mean')

train_a_m = pd.DataFrame(imp_mean.fit_transform(train_a_le))
train_t_m = pd.DataFrame(imp_mean.fit_transform(train_t_le))
train_o_m = pd.DataFrame(imp_mean.fit_transform(train_o_le))
```

* 예측값이 잘 채워진 것을 확인할 수 있음
* 하지만 이 때 모든 값이 NaN인 열은 삭제됨 때문에 PRODUCT_CODE 별로 데이터 프레임이 상이한 형태로 나타남. 이는 후에 각각 모델을 만든다고 해도 test, sub 과 대조에서 문제가 될 수 있고 원본 데이터를 너무 많이 건드리기 때문에 모델링까지 시도하지 않음
